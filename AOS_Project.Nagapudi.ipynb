```python
from tabulate import tabulate
from __future__ import print_function
from datetime import datetime
import psutil
import datetime
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import glob
import pandas as pd
import os
import multiprocessing as mp
import json
import os.path
import re  
import time
from multiprocessing import Process
import threading
import matplotlib.pyplot as plt
import numpy as np

# Performing NB
def NB():
    from sklearn.naive_bayes import BernoulliNB
    from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, log_loss, roc_auc_score
    from sklearn.model_selection import train_test_split

    import pandas as pd
    import numpy as np
    import random

    from sklearn.utils import shuffle
    print("Naive Bayes Classifier Metrics:")

    columns = ['stime', 'sport', 'dport', 'pkts', 'bytes', 'ltime', 'seq', 
            'dur', 'mean', 'sum', 'min', 'max', 'spkts', 'dpkts', 'sbytes', 'dbytes',
            'rate', 'srate', 'drate', 'attack']
    df = pd.read_csv("C:/Users/harsh/Sem2 - TTU/5352-001- Advanced Operating Systems Design/Project/Data_exfiltration.csv", delimiter=';')
    dataset = shuffle(df.filter(columns))

    # input
    x = dataset.iloc[:, [0, 18]].values
  
    # output
    y = dataset.iloc[:, -1].values

    random.seed(121)
    X_train, X_test, Y_train, Y_test = train_test_split(
        x, y, test_size = 0.2, random_state = 42)
    bnb = BernoulliNB()
    bnb.fit(X_train, Y_train)
    Y_pred = bnb.predict(X_test)

    # Calculating Metrics
    cm = confusion_matrix(Y_test, Y_pred)
    ac = accuracy_score(Y_test, Y_pred)
    pr = precision_score(Y_test, Y_pred)
    re = recall_score(Y_test, Y_pred)
    f1 = f1_score(Y_test, Y_pred)
    lolo = log_loss(Y_test, Y_pred)
    rasc = roc_auc_score(Y_test, Y_pred)

    # Displaying Metrics
    # print("Confusion Matrix: {}".format(cm))
    print("Accuracy: {}".format(ac))
    print("Precision: {}".format(pr))
    print("Recall: {}".format(re))
    print("F1 Score: {}".format(f1))
    print("Log Loss: {}".format(lolo))
    print("ROC AUC: {}".format(rasc))

# Performing ANN
def ANN():
    from typing import Type
    from numpy import array
    import tensorflow as tf
    from keras.models import Sequential
    import pandas as pd
    from keras.layers import Dense
    from keras import metrics
    import random
    from tensorflow.python import keras

    from sklearn.utils import shuffle
    print("ANN Classifier Metrics:")

    columns = ['stime', 'sport', 'dport', 'pkts', 'bytes', 'ltime', 'seq', 
            'dur', 'mean', 'sum', 'min', 'max', 'spkts', 'dpkts', 'sbytes', 'dbytes',
            'rate', 'srate', 'drate', 'attack']
    df = pd.read_csv("C:/Users/harsh/Sem2 - TTU/5352-001- Advanced Operating Systems Design/Project/Data_exfiltration.csv", delimiter=';')
    dataset = shuffle(df.filter(columns))

    from sklearn.model_selection import train_test_split
    # input
    x = dataset.iloc[:, [0, 18]].values
  
    # output
    y = dataset.iloc[:, -1].values

    random.seed(121)
    from sklearn.model_selection import train_test_split
    xtrain, xtest, ytrain, ytest = train_test_split(
        x, y, test_size = 0.2, random_state = 0)

    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler().fit(xtrain)
    xtrain = scaler.transform(xtrain)
    xtest = scaler.transform(xtest)  

    from keras.models import Sequential
    from keras.layers import Dense
    model = Sequential()
    model.add(Dense(8, input_dim=2, activation='relu' ))
    model.add(Dense(8, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    from keras import backend as K

    def recall_m(y_true, y_pred):
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
        recall = true_positives / (possible_positives + K.epsilon())
        return recall

    def precision_m(y_true, y_pred):
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
        precision = true_positives / (predicted_positives + K.epsilon())
        return precision

    def f1_m(y_true, y_pred):
        precision = precision_m(y_true, y_pred)
        recall = recall_m(y_true, y_pred)
        return 2*((precision*recall)/(precision+recall+K.epsilon()))

    model.compile(loss='binary_crossentropy',
    optimizer='sgd',
    metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), f1_m, tf.keras.metrics.AUC()])
    model.fit(xtrain, ytrain, epochs=4, batch_size=1, verbose=1)

    y_pred = model.predict(xtest)

    lolo, ac, pr, re, f1, rasc = model.evaluate(xtest, ytest,verbose=1)

    # Displaying Metrics
    # print("Confusion Matrix: {}".format(cm))
    print("Accuracy: {}".format(ac))
    print("Precision: {}".format(pr))
    print("Recall: {}".format(re))
    print("F1 Score: {}".format(f1))
    print("Log Loss: {}".format(lolo))
    print("ROC AUC: {}".format(rasc)) 
    
    sys.stdout.flush()
```
